```python
import pandas as pd

df_2023 = pd.read_csv('/Users/mavinjames/Desktop/Basketball_Modeling/NCAAB_Sports_Reference_Scraper/data/2023/gamelogs_2023.csv')
df_2024 = pd.read_csv('/Users/mavinjames/Desktop/Basketball_Modeling/NCAAB_Sports_Reference_Scraper/data/2024/gamelogs_2024.csv')
df_2025 = pd.read_csv('/Users/mavinjames/Desktop/Basketball_Modeling/NCAAB_Sports_Reference_Scraper/data/2025/gamelogs_2025.csv')
```

```python
df_list = [df_2023, df_2024, df_2025]
all_df = pd.concat(df_list, ignore_index=True)
```

```python
#all_df = all_df.drop(columns = 'Unnamed: 0', axis = 1)
all_df["school_name"] = all_df["school_name"].str.replace(r"NCAA$", "", regex=True)
all_df['date'] = pd.to_datetime(all_df['date'])
```

```python
#is_Home feature
import numpy as np

all_df['game_location'].fillna('', inplace=True)

all_df['is_Home'] = all_df['game_location'].apply(lambda x: 1 if x == '' else (.5 if x == 'N' else 0))
```

```python
#score_diff feature

all_df['score_diff'] = all_df['team_game_score'] - all_df['opp_team_game_score']

#all_df[['score_diff','team_game_score','opp_team_game_score','team_game_result']]
```

```python
#rolling window for 5 and 10 average score differential

all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])["score_diff"]

all_df["avg_score_diff_last_5"] = g.transform(
    lambda s: s.shift(1).rolling(5, min_periods=1).mean()
)

all_df["avg_score_diff_last_10"] = g.transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df[["avg_score_diff_last_5", "avg_score_diff_last_10"]] = (
    all_df[["avg_score_diff_last_5", "avg_score_diff_last_10"]].fillna(0)
)
```

```python
#rolling window for 5 and 10 average points scored
all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])["team_game_score"]

all_df["avg_points_scored_last_5"] = g.transform(
    lambda s: s.shift(1).rolling(5, min_periods=1).mean()
)

all_df["avg_points_scored_last_10"] = g.transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df[["avg_points_scored_last_5", "avg_points_scored_last_10"]] = (
    all_df[["avg_points_scored_last_5", "avg_points_scored_last_10"]].fillna(0)
)
```

```python
#rolling window for 5 and 10 average points scored
all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])["opp_team_game_score"]

all_df["avg_points_given_last_5"] = g.transform(
    lambda s: s.shift(1).rolling(5, min_periods=1).mean()
)

all_df["avg_points_given_last_10"] = g.transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df[["avg_points_given_last_5", "avg_points_given_last_10"]] = (
    all_df[["avg_points_given_last_5", "avg_points_given_last_10"]].fillna(0)
)
```

```python
#rolling window for 5 and 10 average points scored
all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])["team_game_score"]

all_df["avg_points_scored_last_5"] = g.transform(
    lambda s: s.shift(1).rolling(5, min_periods=1).mean()
)

all_df["avg_points_scored_last_10"] = g.transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df[["avg_points_scored_last_5", "avg_points_scored_last_10"]] = (
    all_df[["avg_points_scored_last_5", "avg_points_scored_last_10"]].fillna(0)
)
```

```python
#Rolling win pct for the last 10 games
all_df["win"] = all_df["team_game_result"].map({"W": 1, "L": 0})

g = all_df.groupby(["season", "school_name"])

all_df["win_pct_last_10"] = g["win"].transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df["win_pct_last_10"] = (
    all_df["win_pct_last_10"].fillna(0)
)
```

```python
#Rolling effective field goal percentage last 5 and 10
g = all_df.groupby(["season", "school_name"])

# previous games only
fg_roll_5  = g["fg"].transform(lambda s: s.shift(1).rolling(5, min_periods=1).sum())
fg3_roll_5 = g["fg3"].transform(lambda s: s.shift(1).rolling(5, min_periods=1).sum())
fga_roll_5 = g["fga"].transform(lambda s: s.shift(1).rolling(5, min_periods=1).sum())

all_df["efg_pct_last_5"] = (fg_roll_5 + 0.5 * fg3_roll_5) / fga_roll_5

fg_roll_10  = g["fg"].transform(lambda s: s.shift(1).rolling(10, min_periods=1).sum())
fg3_roll_10 = g["fg3"].transform(lambda s: s.shift(1).rolling(10, min_periods=1).sum())
fga_roll_10 = g["fga"].transform(lambda s: s.shift(1).rolling(10, min_periods=1).sum())

all_df["efg_pct_last_10"] = (fg_roll_10 + 0.5 * fg3_roll_10) / fga_roll_10
```

```python
#rolling window for 5 and 10 average tovs
all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])["tov"]

all_df["avg_tovs_last_5"] = g.transform(
    lambda s: s.shift(1).rolling(5, min_periods=1).mean()
)

all_df["avg_tovs_last_10"] = g.transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df[["avg_tovs_last_5", "avg_tovs_last_10"]] = (
    all_df[["avg_tovs_last_5", "avg_tovs_last_10"]].fillna(0)
)
```

```python
# rolling window for 5 and 10 average ORB
all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])["orb"]

all_df["avg_orb_last_5"] = g.transform(
    lambda s: s.shift(1).rolling(5, min_periods=1).mean()
)

all_df["avg_orb_last_10"] = g.transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

all_df[["avg_orb_last_5", "avg_orb_last_10"]] = (
    all_df[["avg_orb_last_5", "avg_orb_last_10"]].fillna(0)
)
```

```python
#rolling window for 5 and 10 average FTA, AST, and TRB

all_df = all_df.sort_values(["season", "school_name", "date"])

g = all_df.groupby(["season", "school_name"])

for col in ["fta", "ast", "trb"]:
    all_df[f"avg_{col}_last_5"] = g[col].transform(
        lambda s: s.shift(1).rolling(5, min_periods=1).mean()
    )
    all_df[f"avg_{col}_last_10"] = g[col].transform(
        lambda s: s.shift(1).rolling(10, min_periods=1).mean()
    )

all_df[[f"avg_{col}_last_5" for col in ["fta","ast","trb"]] +
       [f"avg_{col}_last_10" for col in ["fta","ast","trb"]]] = (
    all_df[[f"avg_{col}_last_5" for col in ["fta","ast","trb"]] +
           [f"avg_{col}_last_10" for col in ["fta","ast","trb"]]].fillna(0)
)
```

```python
import numpy as np
import pandas as pd

df = merged.copy()

# ----- adjust these if your column names differ -----
POINTS_SCORED_COL = "team_game_score"
POINTS_GIVEN_COL = "opp_team_game_score"
SCORE_DIFF_COL = "score_diff"
TEAM_RESULT_COL = "team_game_result"
FG_COL = "fg"
FG3_COL = "fg3"
FGA_COL = "fga"
# ----------------------------------------------------

# keep raw opponent columns
keep_raw_opp = {"opp_name_abbr", "opp_team_game_score"}

# drop only derived opp_ features
cols_to_drop = [c for c in df.columns if c.startswith("opp_") and c not in keep_raw_opp]
df = df.drop(columns=cols_to_drop, errors="ignore")

# build win if needed
if "win" not in df.columns and TEAM_RESULT_COL in df.columns:
    df["win"] = df[TEAM_RESULT_COL].map({"W": 1, "L": 0})

# sort + group
df = df.sort_values(["season", "school_name", "date"])
g = df.groupby(["season", "school_name"])

def rolling_mean(col, window):
    return g[col].transform(lambda s: s.shift(1).rolling(window, min_periods=1).mean())

# score diff + points
df["avg_score_diff_last_5"]  = rolling_mean(SCORE_DIFF_COL, 5)
df["avg_score_diff_last_10"] = rolling_mean(SCORE_DIFF_COL, 10)

df["avg_points_scored_last_5"]  = rolling_mean(POINTS_SCORED_COL, 5)
df["avg_points_scored_last_10"] = rolling_mean(POINTS_SCORED_COL, 10)

df["avg_points_given_last_5"]  = rolling_mean(POINTS_GIVEN_COL, 5)
df["avg_points_given_last_10"] = rolling_mean(POINTS_GIVEN_COL, 10)

# win pct last 10
df["win_pct_last_10"] = g["win"].transform(
    lambda s: s.shift(1).rolling(10, min_periods=1).mean()
)

# weighted eFG% (sum-based)
fg_roll_5  = g[FG_COL].transform(lambda s: s.shift(1).rolling(5,  min_periods=1).sum())
fg3_roll_5 = g[FG3_COL].transform(lambda s: s.shift(1).rolling(5,  min_periods=1).sum())
fga_roll_5 = g[FGA_COL].transform(lambda s: s.shift(1).rolling(5,  min_periods=1).sum())

df["efg_pct_last_5"] = (fg_roll_5 + 0.5 * fg3_roll_5) / fga_roll_5.replace(0, np.nan)

fg_roll_10  = g[FG_COL].transform(lambda s: s.shift(1).rolling(10, min_periods=1).sum())
fg3_roll_10 = g[FG3_COL].transform(lambda s: s.shift(1).rolling(10, min_periods=1).sum())
fga_roll_10 = g[FGA_COL].transform(lambda s: s.shift(1).rolling(10, min_periods=1).sum())

df["efg_pct_last_10"] = (fg_roll_10 + 0.5 * fg3_roll_10) / fga_roll_10.replace(0, np.nan)

# other rolling stats
for col in ["tov", "orb", "fta", "ast", "trb"]:
    df[f"avg_{col}_last_5"]  = rolling_mean(col, 5)
    df[f"avg_{col}_last_10"] = rolling_mean(col, 10)

# fill rolling NaNs if you want
roll_cols = [
    "avg_score_diff_last_5","avg_score_diff_last_10",
    "avg_points_scored_last_5","avg_points_scored_last_10",
    "avg_points_given_last_5","avg_points_given_last_10",
    "win_pct_last_10","efg_pct_last_5","efg_pct_last_10",
    "avg_tovs_last_5","avg_tovs_last_10",
    "avg_orb_last_5","avg_orb_last_10",
    "avg_fta_last_5","avg_fta_last_10",
    "avg_ast_last_5","avg_ast_last_10",
    "avg_trb_last_5","avg_trb_last_10",
]
df[roll_cols] = df[roll_cols].fillna(0)

# build opponent df with prefix (uses the same game scores)
opp_cols = ["season","date","opp_name_abbr","team_game_score","opp_team_game_score"] + roll_cols + ["win"]
opp_df = df[opp_cols].add_prefix("opp_")

# merge opponent features back onto main df
merged = df.merge(
    opp_df,
    left_on=["season","date","school_name","team_game_score"],
    right_on=["opp_season","opp_date","opp_opp_name_abbr","opp_opp_team_game_score"],
    how="inner"
)
```

```python
cols = [
    "avg_score_diff_last_10",
    "efg_pct_last_10",
    "avg_tovs_last_10",
]

check = merged.merge(
    merged[["date", "school_name"] + cols],
    left_on=["date", "opp_name_abbr"],
    right_on=["date", "school_name"],
    how="left",
    suffixes=("", "_opp_from_self")
)

for c in cols:
    diff = (check[f"opp_{c}"] - check[f"{c}_opp_from_self"]).abs()
    print(c, "mismatches:", (diff > 1e-8).sum(), " / ", diff.notna().sum())
```

```python

cols = ["avg_score_diff_last_10", "efg_pct_last_10", "avg_tovs_last_10"]

check = merged.merge(
    merged[["date", "school_name"] + cols],
    left_on=["date", "opp_name_abbr"],
    right_on=["date", "school_name"],
    how="left",
    suffixes=("", "_opp_from_self")
)

for c in cols:
    mask = (check[f"opp_{c}"] - check[f"{c}_opp_from_self"]).abs() > 1e-8
    print(f"\n{c} mismatches:")
    print(
        check.loc[mask, ["date","school_name","opp_name_abbr", f"opp_{c}", f"{c}_opp_from_self"]]
        .head(30)
    )
```
